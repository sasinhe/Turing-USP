````python
import matplotlib.pyplot as plt

# Dados para cada gráfico, seguindo a nova ordem e com as porcentagens e cores invertidas onde solicitado
dados_graficos = [
    # Gráfico 1: Mesma configuração
    {'probabilidades': [18, 7, 74], 'cores': ['#F9D923', '#CCCCCC', '#EB5353']},

    # Gráfico 2: Era a antiga sexta, com a ordem das porcentagens e cores invertidas
    {'probabilidades': [29, 5, 66], 'cores': ['#36AE7C', '#CCCCCC', '#EB5353']},

    # Gráfico 3: Era a antiga terceira, com a ordem das porcentagens e cores invertidas
    {'probabilidades': [21, 6, 73], 'cores': ['#187498', '#CCCCCC', '#EB5353']},

    # Gráfico 4: Era a antiga quarta, com a ordem das porcentagens e cores invertidas
    {'probabilidades': [33, 11, 56], 'cores': ['#F9D923', '#CCCCCC', '#36AE7C']},

    # Gráfico 5: Era a antiga quinta, com a ordem das porcentagens e cores invertidas
    {'probabilidades': [33, 23, 44], 'cores': ['#187498', '#CCCCCC', '#F9D923']},

    # Gráfico 6: Igual ao antigo segundo gráfico
    {'probabilidades': [16, 13, 70], 'cores': ['#187498', '#CCCCCC', '#36AE7C']}
]

# Labels que serão mostradas no topo do primeiro gráfico
labels = ['GUILHERME BOULOS', 'BRANCO/NULO/INDECISOS', 'TABATA AMARAL']

# Configuração dos gráficos
fig, axs = plt.subplots(nrows=6, ncols=1, figsize=(25, 12))  # 6 gráficos empilhados

# Função para desenhar as barras com porcentagens e cores
def desenhar_grafico(ax, probabilidades, cores, colocar_labels=False):
    left = 0  # Ponto inicial da barra
    soma_prob = sum(probabilidades)

    # Se a soma das probabilidades for 99%, adiciona 0,33333% a cada barra
    if soma_prob == 99:
        probabilidades_ajustadas = [prob + 0.333333 for prob in probabilidades]
    else:
        probabilidades_ajustadas = probabilidades

    # Desenhar as barras com os valores ajustados
    for prob, color in zip(probabilidades_ajustadas, cores):
        ax.barh(0, prob, left=left, color=color, edgecolor='white', height=0.3)
        left += prob  # Atualizando o ponto inicial para a próxima barra

    # Adicionando rótulos das porcentagens no centro de cada segmento (usando valores reais)
    left = 0
    for prob_real in probabilidades:
        ax.text(left + prob_real / 2, 0, f'{prob_real}%', va='center', ha='center', color='white', fontweight='bold', fontsize=20)
        left += prob_real

    # Adicionando os labels no topo apenas para o primeiro gráfico
    if colocar_labels:
        ax.text(0, 0.2, f'     GUILHERME BOULOS        TABATA AMARAL        PABLO MARÇAL        RICARDO NUNES        BRANCO/NULO/INDECISOS', va='center', ha='left', color='#212121', fontweight='bold', fontsize=15)

    # Ajustando os limites e removendo eixos
    ax.set_xlim(0, 100)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)

# Desenhando os 6 gráficos com diferentes porcentagens e cores
for i, (ax, dados) in enumerate(zip(axs, dados_graficos)):
    desenhar_grafico(ax, dados['probabilidades'], dados['cores'], colocar_labels=(i == 0))

# Configurando a cor de fundo da figura
fig.patch.set_facecolor('#ffffff')

# Ajustar o espaçamento entre os gráficos
plt.tight_layout()

# Exibir os gráficos
plt.show()
````
